### The Erdos-Renyi Random Graph model
The version presented here of the Erdos-Renyi random graph model is the G(n,p) model, generating a graph G that has n nodes such that the probability that $ n_{i} $ and $ n_{j} $is exactly p. 

#### Algorithm
The high level algorithm is as follows:
1. Add all (n) nodes
2. $\forall i, j$ such that $ i \neq j $, generate random number $r \in [0,1]$
3. If r <= p, add an edge between i and j, otherwise pass

The time complexity of the above algorithm is very clearly $ O(n^{2}) $, operating once per pair $ i,j$ of nodes

#### Probability Analysis
Suppose we are interested in a particular graph G with n nodes and m edges. What is the probability that Erdos-Reyni generates exactly G? 

Observe that at each generation of random number r, we are essentially flipping a biased coin, with probability $p$ to land heads (to add an edge), and probability $1-p$ to land tails (pass). For graph G with n nodes, define M as the total number of possible edges, which is $\binom{n}{2}$. Thus, find that the total probability to gather graph G is:
$$ P(G) = \binom{M}{m}p^{m}(1-p)^{M-m} $$

Out of total M possible edges, we choose m to exist, with p probability of each one existing, and (1-p) probability of (M-m) of them not existing.
